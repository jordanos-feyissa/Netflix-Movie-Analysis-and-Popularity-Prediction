# Classification

 - Classify based on salary the experience level

## Data preparation

### Drop unecessary columns

```python
len(df[df["company_location"]==df["employee_residence"]])
```

Employee residence is redundant. Out of 15,000 entries, 10,744 employees live in the same location as their company.

```python
columns_to_drop = ['job_id' ,  'salary_currency', 'posting_date', 'application_deadline', 'job_description_length',  'employee_residence',  'required_skills', 'job_title' , 'education_required', "years_experience" , 'company_name']
df_ = df.drop(columns = columns_to_drop )
```

## XGBoost

#### split data into train and test set

```python
X = df_.drop(columns=['experience_level'])
y= df_['experience_level']

X_train , X_test, y_train, y_test = train_test_split( X , y, test_size=0.3,  random_state=42 , stratify=y)                
```

```python
#applying target encoding
from category_encoders import TargetEncoder

cat_cols = [ 'company_location', 'industry'] #'job_title' 

encoder = TargetEncoder(cols = cat_cols)

X_train = encoder.fit_transform(X_train, y_train)
X_test = encoder.transform(X_test)
```

```python
#applying Label encoding
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ['company_size']

categories = [ ["S", "M", "L"]]
        
             

encoder = OrdinalEncoder(categories= categories)  # define order

X_train[ordinal_cols] = encoder.fit_transform(X_train[ordinal_cols] )
X_test[ordinal_cols] = encoder.transform(X_test[ordinal_cols])
```

```python
#One hot encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

X_train_ohe = ohe.fit_transform(X_train[["employment_type"]])
X_test_ohe = ohe.fit_transform(X_test[["employment_type"]])
```

```python
#Convert array into dataframe
ohe_cols = ohe.get_feature_names_out(["employment_type"])

X_train_ohe = pd.DataFrame(X_train_ohe, columns=ohe_cols, index=X_train.index)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=ohe_cols, index=X_test.index)
```

```python
# Drop original categorical column

X_train = X_train.drop(columns=["employment_type"])
X_test = X_test.drop(columns=["employment_type"])
```

```python
#concatenate the encoded column 
X_train = pd.concat([X_train, X_train_ohe], axis=1)
X_test = pd.concat([X_test , X_test_ohe], axis=1)
```

```python
#Encoding the categorical column
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()

y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

```python
### Train the model
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train_enc)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test_enc, y_pred))
print(classification_report(y_test_enc , y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test_enc, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm , annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
sns.boxplot(data=X_train.join(y_train), x=y_train, y="salary_usd")
plt.title("Salary distribution by entry level")
plt.show()
```

Analysis of the dataset shows substantial salary overlap across seniority levels, mainly due to differences in company size and industry. Larger organizations tend to pay higher salaries at all levels, making salary alone an unreliable indicator of seniority.

Despite this overlap and the absence of explicit seniority-related features, the model achieved an accuracy of 84%, indicating strong predictive performance. 

The model’s main source of misclassification occurs among senior-level employees, who frequently overlap with both junior and expert categories. Similarly, expert-level employees are often confused with senior and, to a lesser extent, junior employees. This pattern is mainly caused by the strong overlap in salary ranges and employment characteristics between adjacent seniority levels, making it difficult for the model to identify clear boundaries between them.

## Catboost

```python
columns_to_drop = ['job_id' , 'salary_currency',  'posting_date', 'application_deadline', 'job_description_length',  'employee_residence',  'required_skills',  'education_required', "years_experience" , 'company_name']
df_ = df.drop(columns = columns_to_drop )
```

```python
X = df_.drop(columns=['experience_level'])
y= df_['experience_level']

X_train , X_test, y_train, y_test = train_test_split( X , y, test_size=0.3,  random_state=42 , stratify=y)                
```

```python
cat_cols = ['job_title', 'employment_type', 'company_location', 'company_size',  'industry']

from catboost import CatBoostClassifier

model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function="MultiClass",
    eval_metric="Accuracy",
    verbose=200
)

model.fit(
    X_train,
    y_train, 
    cat_features=cat_cols,
    eval_set=(X_test, y_test)
)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, preds)
print("CatBoost accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute confusion matrix
cm = confusion_matrix(y_test, preds)

# Display it
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix - CatBoost")
plt.show()
```

CatBoost achieved an accuracy of 84%, matching the performance obtained with XGBoost. The confusion matrix shows that Senior employees are frequently misclassified as either Mid-level or Expert, while Expert employees are often predicted as Mid-level or Senior. This pattern reflects the substantial overlap in salary ranges across industries and company sizes, which makes the separation between adjacent seniority levels difficult. As a result, the model tends to confuse neighboring classes, suggesting that most misclassifications are driven by the inherent overlap in the data rather than by limitations in the model itself.

# Regression

- Predict Salary

```python
columns_to_drop = ['job_id', 'posting_date', 'application_deadline', 'job_description_length',  "years_experience",  'required_skills', 
                   'employee_residence' ]
df_ = df.drop(columns = columns_to_drop )
```

```python
df_["log_salary"] = np.log(df_["salary_usd"])
```

```python
X = df_.drop(columns=["salary_usd"])
y= df_["salary_usd"]

X_train , X_test, y_train, y_test = train_test_split( X , y, test_size=0.3,  random_state=42 )    
```

```python
from catboost import CatBoostRegressor

cat_cols = ['job_title', 'experience_level','employment_type', 'company_location', 'company_size',  'industry', 'company_name', 
            'salary_currency', 'education_required']


model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function="RMSE",
    eval_metric="RMSE",
    verbose=200
)

model.fit(
    X_train,
    y_train,
    cat_features=cat_cols,
    eval_set=(X_test, y_test)
)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predictions
y_pred = model.predict(X_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("MAE:", mae)
print("R²:", r2)
```

The model achieves an RMSE of 19,927 USD and an MAE of 14,335 USD, which are reasonable given the substantial salary variation across countries, industries, company sizes, and seniority levels. The R² score of 0.892 indicates that the model explains nearly 89% of the variance in salary, confirming strong predictive performance despite the inherent complexity and noise present in real-world compensation data.

```python
plt.scatter(range(len(y_test)), y_test, color="green", alpha=0.5, label="Actual Salary")
plt.scatter(range(len(y_test)), y_pred, color="purple", alpha=0.5, label="Predicted Salary")

plt.xlabel("Sample Index")
plt.ylabel("Salary")
plt.title("Actual vs Predicted Salary")
plt.legend()
plt.show()
```

```python
errors = abs(y_test - y_pred)

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, c=errors, cmap="coolwarm", alpha=0.6)
plt.colorbar(label="Absolute Error")

plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'k--', label="Perfect Prediction")

plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary (Colored by Error)")
plt.legend()
plt.tight_layout()
plt.show()
```

Although the model tends to compress extreme values (underpredicting the highest salaries and slightly overpredicting the lowest), its overall predictive accuracy remains high and consistent with expectations for real‑world salary data.

```python
feature_importance = model.get_feature_importance()
features = X_train.columns

fi_df = pd.DataFrame({
    "feature": features,
    "importance": feature_importance
}).sort_values(by="importance", ascending=False)

print(fi_df)
```

The feature-importance results show that salary is primarily driven by experience level and company location, which dominate the model’s predictive power. Company size also contributes to the prediction, but to a much lesser extent. All remaining features, including job title, education, industry, benefits, remote ratio, and employment type, have only marginal individual impact on the model’s predictions.
