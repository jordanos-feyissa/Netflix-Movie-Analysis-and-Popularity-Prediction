## Exploratory data analyis

#### Salary by Year

```python
df.groupby("work_year")["salary_in_usd"].describe()
```

#### Salary by Experience

```python
df.groupby("experience_level")["salary_in_usd"].describe()
```

#### Salary by Employment Type

```python
df.groupby("employment_type")["salary_in_usd"].describe()
```

```python
# chek distribution of employement type
df["employment_type"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(7,7), pctdistance= 0.8 , labeldistance=1.1
                                              ,title="Employment Type Distribution");
```

```python
#Job title distribution
df["job_title"].value_counts().plot.bar(figsize=(18,10), fontsize=14);
```

```python
#check all company location

import squarify

counts = df["company_location"].value_counts()
squarify.plot(sizes=counts.values, label=counts.index, alpha=0.8)
```

```python
#check all company location

import squarify

counts = df["employee_residence"].value_counts()
squarify.plot(sizes=counts.values, label=counts.index, alpha=0.8)
```

```python
df["remote_ratio"].value_counts().sort_index().plot.bar(figsize=(12,6));
#0 No remote work (less than 20%) 50 Partially remote 100 Fully remote (more than 80%)
```

```python
#Check the distribution of company size
df["company_size"].value_counts().sort_index().plot.bar(figsize=(12,6));
#S less than 50 employees (small) M 50 to 250 employees (medium) L more than 250 employees (large)
```

#### Salary Distribution

```python
#Check the distribution of salary

sns.histplot(df["salary_in_usd"], bins="fd", kde=True)
plt.title("Salary Distribution (USD)")
plt.show()
             
```
